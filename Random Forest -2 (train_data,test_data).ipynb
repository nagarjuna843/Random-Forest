```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree 
df_train=pd.read_csv('C:/Users/PC/Downloads/train.csv')
df_test=pd.read_csv('C:/Users/PC/Downloads/test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
input_train=df_train.drop(['PassengerId','Name','Ticket','Cabin'],axis='columns',inplace=True)
input_test=df_test.drop(['PassengerId','Name','Ticket','Cabin'],axis='columns',inplace=True)
```

```python
input_train = df_train.drop('Survived',axis='columns')
target_train = df_train.Survived
```

```python
input_train.head()
```

```python
input_test=df_test
input_test.head()
```

```python
input_test['Fare'].isnull().value_counts()
```

```python
input_train.Age = input_train.Age.fillna(input_train.Age.mean())
input_train.Embarked = input_train.Embarked.fillna('S')
input_test.Age = input_test.Age.fillna(input_test.Age.mean())
input_test.Fare = input_test.Fare.fillna(input_test.Fare.mean())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
input_train['Pclass_n']=LabelEncoder().fit_transform(input_train['Pclass'])
input_train['Sex_n']=LabelEncoder().fit_transform(input_train['Sex'])
input_train['Age_n']=LabelEncoder().fit_transform(input_train['Age'])
input_train['SibSp_n']=LabelEncoder().fit_transform(input_train['SibSp'])
input_train['Parch_n']=LabelEncoder().fit_transform(input_train['Parch'])
input_train['Fare_n']=LabelEncoder().fit_transform(input_train['Fare'])
input_train['Embarked_n']=LabelEncoder().fit_transform(input_train['Embarked'])
```

```python
input_test['Pclass_n']=LabelEncoder().fit_transform(input_test['Pclass'])
input_test['Sex_n']=LabelEncoder().fit_transform(input_test['Sex'])
input_test['Age_n']=LabelEncoder().fit_transform(input_test['Age'])
input_test['SibSp_n']=LabelEncoder().fit_transform(input_test['SibSp'])
input_test['Parch_n']=LabelEncoder().fit_transform(input_test['Parch'])
input_test['Fare_n']=LabelEncoder().fit_transform(input_test['Fare'])
input_test['Embarked_n']=LabelEncoder().fit_transform(input_test['Embarked'])
```

```python
inputs_n=input_train.drop(['Pclass','Sex','Age','SibSp','Fare','Embarked'],axis='columns')
inputs_e=input_test.drop(['Pclass','Sex','Age','SibSp','Fare','Embarked'],axis='columns')
```

```python
inputs_e
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=200)
model.fit(inputs_n,target_train)
```

```python
model.score(inputs_n,target_train)
```

```python
predict=model.predict(inputs_e)
predict
```
